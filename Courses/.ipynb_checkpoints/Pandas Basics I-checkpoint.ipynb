# Label Based Indexing:

## 1. Selecting Rows with loc[]:

```python
summer.loc["HAJOS, Alfred"]
```

```python
summer.loc["PHELPS, Michael"]
```

## 2. Slicing Rows and Columns with loc[]:

```python
summer.loc["PHELPS, Michael", ["Medal", "Event"]]
```

```python
summer.loc[["PHELPS, Michael", "HOFMANN, Fritz"], ["Medal", "Event"]]
```

```python
summer.loc[:"CHASAPIS, Spiridon"]
```

##### Error when slicing with non unique label 

```python
summer.loc[:"PHELPS, Michael"]
```

```python
summer.head(20)
```

```python
summer.loc["DRIVAS, Dimitrios":"BLAKE, Arthur"]
```

#### Error in case we specify non-existent column

```python
 summer.loc["PHELPS, Michael": "Kind"]
```
